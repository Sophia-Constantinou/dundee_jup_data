```python
# There's something wrong with this - I get huge sequences
# All of them are the same?? - fixed it

data = pd.read_csv('merged_final.csv')


for i,row in data.iterrows():
    data_pdb = []
    data_unip = []
       
    pdb_id = row.iloc[2]
#     print(pdb_id)
    pdb_seq = row.iloc[4] 
#     print(len(pdb_seq))
    uniprot_seq = row.iloc[6]
    scop = row.iloc[1]
#     print(scop)

    # Find chain used from SCOPe ID
    try:
        chain = scop[-2]
        chain_cap = chain.upper()   

        df3 = pd.read_csv('/cluster/gjb_lab/2524591/data/SIFTS/%s.csv' % pdb_id)
        
        df3_null = df3.replace('null', np.nan, regex=True)
        df3_val = df3_null.dropna()
#     print(df3_val)
#         print(df3_val)
    


    # PDB
        pdb_offset = df3_val.iloc[:, [1,2]]
        pdb_off_str = pdb_offset.astype(str)

        # Choose chain using SCOPe ID
        df_chain = pdb_off_str.loc[pdb_off_str['PDB_dbChainId'].str.startswith(chain_cap)]
#     print(df_chain)      
                                  
        df_res_pdb = df_chain.iloc[:,1]
    
        # Turn string into int
        try:
            df_fl = df_res_pdb.astype(float)
            pdb_int = df_fl.astype(int)

        # If it's not an integer - sometimes there's letters
        except ValueError:
            print(f'Skipping {pdb_id}: PDB - Invalid data...')
            continue


        df_p = pd.DataFrame(pdb_int)
#     print(df_p)
    
#         first_val_row = df_p.first_valid_index()
#         print(df_p, first_val_row)
        last_loc = df_p.iloc[-1]
        last_row = last_loc.name
#         print(last_loc, last_row)
    
    # SCOPe ID has different configurations - chain might be in last letter, but in most it's in second to last
    except IndexError:
        print(f'SCOPe ID not valid: PDB_ID = {pdb_id}')
        continue

    # Iterate over all rows
    for row in df_p.iterrows():
        num = row[1][0]
#         print(num)
        
        if num < len(pdb_seq):
            pdbs = pdb_seq[num]
            data_pdb.append(pdbs)

        else:
            print("Too large: " + pdb_id)
            break
#         print(len(data_pdb))
                                          
    pdb_d = pd.DataFrame(data_pdb)
#     print(pdb_d)
        
    pdb_t = pdb_d.T
    pdb_t['%s' % pdb_id] = pdb_t.apply(lambda x: ''.join(x.dropna().astype(str)),axis=1)
    pdb_fin = pdb_t['%s' % pdb_id]
#     print(pdb_fin)
       
    pdb_fin.to_csv('/homes/2524591/data/pdb_merged/%s.csv' % pdb_id)
   
    
      
    # Uniprot -- start from same row as pdb - use first_val_row
     
    uniprot_offset = df3_val.iloc[:last_row, 4]
#     print(df3_val, first_val_row, last_row, uniprot_offset)
    try:
#         u_fl = uniprot_offset.astype(float)
        u_int = uniprot_offset.astype(int)
    except ValueError:
        print(f'Skipping {pdb_id}: Uniprot - Invalid data...')
        continue
                    
    df_u = pd.DataFrame(u_int)

        
    for row in df_u.iterrows():
        y = row[1][0]
        
        if y < len(uniprot_seq):
            uni_final = uniprot_seq[y]
            data_unip.append(uni_final)
            
        else:
            print("Too large: " + pdb_id)
            break
                
    uni_d = pd.DataFrame(data_unip)
#     print(df_p, df_u, pdb_d, uni_d)
    uni_t = uni_d.T
    uni_t['%s' % pdb_id] = uni_t.apply(lambda x: ''.join(x.dropna().astype(str)),axis=1)
    uni_fin = uni_t['%s' % pdb_id]
#     print(uni_fin)
#     print(pdb_fin)
#     add_pdb_uni = uni_fin.iloc[[0]]
              
    uni_fin.to_csv('/homes/2524591/data/uni_merged/%s.csv' % pdb_id)
```
